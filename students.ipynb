## Predicting student's marks


We have used LinearRegressor and Support Vector Regressor for predictions

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('student-mat.csv', sep=';')
```

```python
data.head()
```

```python
data.columns
```

### Defining features and target data

```python
X = data[['G1', 'G2', 'failures', 'absences', 'studytime']]
y = data['G3']
```

### Splitting data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

### Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
```

#### Analyzing the predictions

```python
#from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Accuracy
acc = linreg.score(X_test, y_test)
acc
```

```python
y_test.iloc[3]
```

```python
# printing the predictions with the used features

for i in range(len(y_pred)):
    print('Predicted Marks:', y_pred[i])
    print('Actual G3 Marks:', y_test.iloc[i],'\n')
```

### Appliying Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel='linear')
```

```python
svr.fit(X_train, y_train)
```

```python
# prediction

svr_y_pred = svr.predict(X_test)
```

```python
# Testing Accuracy

svr.score(X_test, svr_y_pred)
```

```python
# Accuracy in Training set

svr.score(X_train, y_train)
```

```python
#X_test.iloc[0]
```

```python
# printing the predictions with the used features

for i in range(len(svr_y_pred)):
    print('Predicted Marks:', svr_y_pred[i])
    print('Actual G3 Marks:', y_test.iloc[i])
    print('Other features were', X_test.iloc[i].to_list(),'\n')
```

### Conclusion
**We can say that SVR outplayed LinearRegressor**
